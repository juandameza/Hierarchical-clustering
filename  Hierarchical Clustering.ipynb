Agrupamiento en Hierarchical Clustering


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score,davies_bouldin_score
from sklearn.cluster import AgglomerativeClustering
#Librería para evitar la advertencia de que el modelo tiene una pérdida de memoria en Windows 
import warnings
warnings.filterwarnings('ignore')
```

```python
# Cargar el dataset
data = pd.read_csv('Mall_Customers.csv') 

```

```python
# Realizamos lectura de los Datos para el análisis exploratorio
data.head(40)
```

```python
# Descripción de los datos
data.describe()
```

```python
data.isnull().sum()
```

```python
sns.pairplot(data)
plt.show()
```

```python
data.dropna(inplace=True)  # Eliminar filas con valores nulos
```

```python
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
selected_features = scaled_data  # Utilizamos las características escaladas
```

```python
linked = linkage(selected_features, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(linked, truncate_mode='lastp', p=10)
plt.title('Dendrograma')
plt.xlabel('Clientes')
plt.ylabel('Distancia Euclidiana')
plt.show()
```

```python
n_clusters = 5  # Número de clusters óptimo basado en el dendrograma
model = AgglomerativeClustering(n_clusters=n_clusters, metric='euclidean', linkage='ward')
cluster_labels = model.fit_predict(selected_features)
data['Cluster'] = cluster_labels
```

```python
silhouette_avg = silhouette_score(selected_features, cluster_labels)
calinski_harabasz = calinski_harabasz_score(selected_features, cluster_labels)
db_score = davies_bouldin_score(scaled_data, cluster_labels)

print(f"Coeficiente de Silhouette: {silhouette_avg}")
print(f"Índice de Calinski-Harabasz: {calinski_harabasz}")
print(f"Índice de Davies-Bouldin: {db_score}")
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x=data['Annual Income (k$)'], 
    y=data['Spending Score (1-100)'], 
    hue=data['Cluster'], 
    palette='viridis'
)
plt.title('Segmentación de Clientes (Clusters)')
plt.xlabel('Ingreso Anual (k$)')
plt.ylabel('Puntaje de Gasto (1-100)')
plt.show()
```

```python
sns.countplot(x='Cluster', data=data, palette='viridis')
plt.title('Distribución de Clientes por Cluster')
plt.show()
```

```python
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x=data['Annual Income (k$)'], 
    y=data['Spending Score (1-100)'], 
    hue=data['Cluster'], 
    palette='viridis'
)
plt.title('Segmentación de Clientes (Clusters)')
plt.xlabel('Ingreso Anual (k$)')
plt.ylabel('Puntaje de Gasto (1-100)')
plt.show()
```

Interpretar, analizar y documentar los resultados obtenidos

El objetivo fue realizar una segmentación de clientes utilizando clustering jerárquico basado en el dataset de "Mall Customer Segmentation". Se emplearon tres métricas clave para evaluar la calidad del modelo: Coeficiente de Silhouette, Índice de Calinski-Harabasz e Índice de Davies-Bouldin.

-Número de Clusters: 5 (determinado mediante análisis del dendrograma).

-Características utilizadas: Ingreso Anual (Annual Income) y Puntaje de Gasto (Spending Score).

Coeficiente de Silhouette: 0.5538

Indica que los clusters están bien separados y que los puntos dentro de cada cluster están razonablemente cerca del centroide. Este valor sugiere que la segmentación es adecuada.

Un valor mayor a 0.5 es considerado bueno en clustering no supervisado, lo que respalda que los grupos identificados tienen sentido.

Índice de Calinski-Harabasz: 244.41

Este índice mide la relación entre la dispersión intra-cluster (compactación) y la inter-cluster (separación). Un valor alto, como este, indica clusters compactos y bien diferenciados. Para este dataset, 244.41 es un resultado sólido que respalda una segmentación clara.


Índice de Davies-Bouldin: 0.5778

Un valor más cercano a 0 indica clusters bien definidos con poco solapamiento. Con un valor de 0.5778, los grupos tienen una separación considerable y están bien delimitados.Este resultado refuerza que los clusters actuales son útiles para segmentación práctica.


Análisis por Clusters

Se identificaron cinco clusters en función de los datos procesados. Los clusters representan diferentes perfiles de clientes según su capacidad de ingreso y su comportamiento de gasto.

1	Bajo	Bajo	Clientes con baja capacidad adquisitiva y gasto.

2	Medio	Alto	Clientes moderados con alta inclinación al gasto.

3	Alto	Bajo	Clientes con alta capacidad, pero bajo consumo.

4	Alto	Alto	Clientes premium: alto ingreso y alta disposición.

5	Bajo	Alto	Clientes jóvenes u ocasionales con gastos elevados.

Cluster 1:
Clientes de menor ingreso con bajo gasto.
Probablemente más sensibles al precio o con poca frecuencia de visitas.
Estrategia: Promociones de bajo costo o eventos locales para atraerlos.

Cluster 2:
Ingreso moderado con alto gasto.
Clientes leales con alto potencial de conversión en embajadores de marca.
Estrategia: Ofrecer programas de fidelización.

Cluster 3:
Clientes con ingresos altos, pero gasto bajo.
Representan un desafío porque tienen capacidad adquisitiva, pero no consumen.
Estrategia: Incentivos específicos para productos premium.

Cluster 4:
Clientes de alto valor con altos ingresos y gastos.
Representan el núcleo de consumidores importantes para el mall.
Estrategia: Personalización, membresías VIP y experiencias exclusivas.

Cluster 5:
Clientes de ingresos bajos, pero con alto gasto.
Pueden ser jóvenes o compradores impulsivos.
Estrategia: Crear experiencias que fortalezcan la lealtad y el engagement.
